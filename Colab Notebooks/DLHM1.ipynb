```
import numpy as np
df['X4']=1.0 # add intercept
X = df[["X1", "X2", "X3","X4","X5"]].to_numpy()
Y = df[["Y"]].to_numpy()
```

```
print(X.shape,Y.shape)
```

```
X = X.astype('double')
```

```
w = tf.Variable(initial_value=tf.random.normal(shape=(5,1),stddev=.01,dtype=tf.float64))
```

```
print(w)
```

```
def poisson_regression(x,W):
    return tf.exp(tf.matmul(X, w))
```

```
def loss_function(y_pred, y_true):

  return tf.reduce_mean(-y_true*tf.math.log(y_pred)+y_pred)
```

```
optimizer = tf.optimizers.SGD(.005)
for epoch in range(15000):
    with tf.GradientTape() as tape:
        y_predicted = poisson_regression(X,w)
        loss = loss_function(y_predicted, Y)
        #loss = tf.keras.losses.poisson(Y, y_predicted)

  
    gradients = tape.gradient(loss, [w])

    optimizer.apply_gradients(zip(gradients,[w]))

print(w)
print(loss)
```
